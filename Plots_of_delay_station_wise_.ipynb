```python
import pandas as pd 
import numpy as np 
import time
import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corrected_train_data (2).csv')
stations_U = ['HCC-U','IFOC-U','MGRO-U','SKRP-U','GE-U','AJG-U','GTNI-U','SLTP-U','CHTP-U','QM-U','SAKT-U','MVNR-U','HKS-U','GNKP-U','AIIM-U','INA-U','JB-U','RCC-U','UDB-U','CTST-U','PTCK-U','CNPE-U','NDI-U','CWBR-U','DEM-U','KG-U','CL-U','VS-U','VW-U','GTBR-U','MDTW-U','AZU-U','AHNR-U','JGPI-U','BIMR-U','RISE-U','SPBI-U']
```

```python
df['ActualDwellTime'] = df['ActualDwellTime'].str.replace(',','')
df['ActualDwellTime'] = df['ActualDwellTime'].fillna(-1)
df['ActualDwellTime'] = df['ActualDwellTime'].astype('float')
df['ActualDwellTime'] = df['ActualDwellTime'].astype('int64')
```

```python
df
```

```python
initial  = '7:30:00'
final = '12:30:00'
ini_cov = datetime.datetime.strptime(initial, '%H:%M:%S')
fin_cov = datetime.datetime.strptime(final, '%H:%M:%S')
```

```python
df
```

```python

for station in stations_U :
  list1 = []
  list2 = []
  for i in range(len(df)):
    if(df['ArrivalTime'][i] == '' or df['Platform'][i]!=station):
      continue
    temp = df['ArrivalTime'][i]
    temp_cov = datetime.datetime.strptime(temp, '%H:%M:%S')
    if temp_cov>=ini_cov and temp_cov<=fin_cov :
      list1.append(temp)
      list2.append(df['ActualDwellTime'][i])
 
  plt.plot(list1, list2)
  plt.xlabel('ArrivalTime')
  plt.ylabel('ActualDwellTime')
  plt.title(station)
  plt.show()
    
```
